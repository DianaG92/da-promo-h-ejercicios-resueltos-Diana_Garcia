```python
df_copia = df.copy()
```

```python
# Eliminamos las columnas que tienen la misma información.
df_copia.drop(["static_fire_date_unix", "date_local", "date_unix", "flight_number"], axis=1, inplace=True)
```

```python
# Sustituimos los valores de 0 y 1 en la columna "success", para que nos den información de mayor valor.
mapa = {0: "Fracaso", 1: "Exito"}

df_copia["success"] = df_copia["success"].map(mapa)

df_copia["success"].unique()
```

```python
# Convertimos a minutos los segundos, en la columna de "window"
df_copia["window"] = df_copia["window"].apply(lambda x: x / 60)

df_copia.sample(1)
```

```python
# Descomprimimos la columna "fairings" para acceder a los datos del diccionario.
df_copia[["reused", "recovery_attempt", "recovered"]] = df_copia["fairings"].apply(pd.Series).iloc[:, :3]

df_copia.head(1)
```

```python
# Descomprimimos la columna "links" para acceder a los datos del diccionario.
df_copia[["webcast", "article"]] = df_copia["links"].apply(pd.Series)[["webcast", "article"]]

df_copia.head(1)
```

```python
# Descomprimimos la columna "failures" para acceder a los datos del diccionario.
df_copia[["time", "altitude", "reason"]] = df_copia["failures"].apply(pd.Series)[0].apply(pd.Series).iloc[:, :3]

df_copia.head(1)
```

```python
# Eliminamos las columnas ("fairings", "links", "failures") que hemos descomprimido.
df_copia.drop(["fairings", "links", "failures"], axis=1, inplace=True)
df_copia.sample(2)
```

```python
# Convertimos la columna "static_fire_date_utc" a formato datetime.
df_copia["static_fire_date_utc" ] = pd.to_datetime(df_copia["static_fire_date_utc"])
df_copia.head(1)
```

```python
# Creamos dos nuevas columnas: "year", "month".
df_copia[["year", "month"]] = df_copia["static_fire_date_utc"].apply(lambda x: pd.Series([x.year, x.month]))
df_copia.head(1)
```

```python
columnas = df_copia.columns

for i in columnas:
    
    # reemplazamos los nulos por el valor Unknown para cada una de las columnas de la lista
    df_copia[i] = df_copia[i].fillna("Unknown")
    
# comprobamos si quedan nulos en las columnas categóricas. 
print("Después del reemplazo usando 'fillna' quedan los siguientes nulos:")
df_copia[columnas].isnull().sum()
```

```python
# Averiguamos el año en el que se lanzaron mas cohetes utilizando .groupby()
df_copia.groupby(["year"])["id"].count().reset_index()
```

```python
# Averiguamos el año en el que se lanzaron mas cohetes utilizando .groupby()
df_copia.groupby(["month"])["id"].count().reset_index()
```

```python
# Averiguamos los lanzamientos que hubo con éxito utilizando .groupby()
df_copia.groupby(["success"])["id"].count().reset_index()
```

```python
# Averiguamos cuál es la ventana de duración aplicando diferentes métodos y agrupándolo por año.
# Primero tenemos que convertir la columna "window" a tipo numérico.
df_copia["window"] = pd.to_numeric(df_copia["window"], errors='coerce')
df_copia.groupby("year")["window"].agg(["mean", "median", "max", "min", "count"]).round(2)
```
