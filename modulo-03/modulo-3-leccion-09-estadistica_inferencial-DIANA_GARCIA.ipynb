2. Ejercicios intervalo de confianza: Para cada uno de los ejercicios interpreta los resultados y piensa qué ventaja a nivel de negocio podemos sacar de calcularlo. 

- Trabajas para una plataforma de comercio en línea y deseas determinar el intervalo de confianza al 95% para el precio promedio de productos con una cantidad de media de valoraciones al producto de 500.

```python
# Calculamos el precio promedio de los productos 
media_precio = df["price"].mean()

# Calculamos el error estándar utilizando el método "sem" de la librería stats
error_precio = stats.sem(df["price"])

# Definimos el nivel de confianza (95% en este caso)
nivel_confianza = 0.95

# Calculamos los grados de libertad de la muestra. Debemos restar el total de datos que tenemos -1. 
grados_libertad = len(df["price"]) - 1

# Calculamos el valor crítico de la distribución de la muestra
valor_critico = stats.t.ppf((1 + nivel_confianza) / 2, df = grados_libertad)

# Calculamos el intervalo de confianza
limite_inferior = media_precio - valor_critico * error_precio
limite_superior = media_precio + valor_critico * error_precio

print("Intervalo de confianza para el precio promedio de los productos:")
print(f"Media Muestral: {np.round(media_precio, 2)}")
print(f"Error Estándar: {np.round(error_precio, 2)}")
print(f"Nivel de Confianza: {nivel_confianza}")
print(f"Valor Crítico: {np.round(valor_critico, 2)}")
print(f"Intervalo de Confianza: ({np.round(limite_inferior, 2)}, {np.round(limite_superior, 2)})")
```

- Trabajas en la gestión de inventario de una tienda en línea y necesitas calcular el intervalo de confianza al 99% para el precio promedio de productos según su país de origen.

```python
# Agrupamos los precios por país de origen
precios_pais = [df[df["origin_country"] == pais]["price"] for pais in df["origin_country"].unique()]

# Calculamos el precio promedio para cada país
media_pais = [i.mean() for i in precios_pais]

# Calculamos el error estándar para cada país
error_pais = [stats.sem(grupo) for grupo in precios_pais]

# Definimos el nivel de confianza (99% en este caso)
nivel_confianza = 0.99

# Calculamos el número de datos en cada grupo
datos = [len(i) for i in precios_pais]

# Calculamos los grados de libertad total
grados_libertad = sum(datos) - len(precios_pais)

# Calculamos el valor crítico de la distribución de la muestra
valor_critico = stats.t.ppf((1 + nivel_confianza) / 2, df = grados_libertad)

# Calculamos el intervalo de confianza para cada país
intervalos_confianza = [
    (media - valor_critico * error, media + valor_critico * error)
    for media, error in zip(media_pais, error_pais)
]

# Imprimimos los intervalos de confianza para cada país
for i, pais in enumerate(df["origin_country"].unique()):
    print(f"Intervalo de confianza al 99% para el precio promedio en {pais}:")
    print(f"Media Muestral: {np.round(media_pais[i], 2)}")
    print(f"Error Estándar: {np.round(error_pais[i], 2)}")
    print(f"Número de datos: {datos[i]}")
    print(f"Valor Crítico: {np.round(valor_critico, 2)}")
    print(f"Intervalo de Confianza: {np.round(intervalos_confianza[i][0], 2)} - {np.round(intervalos_confianza[i][1], 2)}")
    print()
```

- Eres un gerente de ventas en una plataforma de comercio electrónico y quieres determinar el intervalo de confianza al 95% para la valoración promedio de los vendedores según si ofrecen envío express o no.

```python
# Creamos una función para calcular los intervalos de confianza
def intervalo_confianza(dataframe, columna, nivel_confianza = float):
    media = dataframe[columna].mean()
    error = stats.sem(dataframe[columna])

    # Calculamos los grados de libertad de la muestra
    grados_libertad = len(df[columna]) - 1

    # Calculamos el valor crítico (utilizando la distribución t de Student)
    valor_critico = stats.t.ppf((1 + nivel_confianza) / 2, df=grados_libertad)

    # Calculamos el intervalo de confianza
    limite_inferior = media - valor_critico * error
    limite_superior = media + valor_critico * error

    print(f"Intervalo de Confianza para {columna}:")
    print(f"Media Muestral: {np.round(media, 2)}")
    print(f"Error Estándar: {np.round(error, 2)}")
    print(f"Nivel de Confianza: {nivel_confianza}")
    print(f"Valor Crítico: {np.round(valor_critico, 2)}")
    print(f"Intervalo de Confianza: ({np.round(limite_inferior, 2)}, {np.round(limite_superior, 2)})")
```

3. Ejercicios prueba de hipótesis:

- Prueba de Hipótesis sobre el precio promedio de productos con o sin empleados calificados

   - *Contexto:* Trabajas en una plataforma de comercio en línea que ofrece productos de moda. Quieres saber si hay una diferencia significativa en el precio promedio de productos entre aquellos vendidos por empleados calificados (rating_employee > 4) y aquellos vendidos por empleados menos calificados (rating_employee <= 4). Esto te ayudará a tomar decisiones sobre la estrategia de precios y capacitación de empleados.

   - *Hipótesis Nula (H0):* No hay diferencia significativa en el precio promedio de productos entre empleados con buena y mala reputación. 
      
   - *Hipótesis Alternativa (H1):* Existe una diferencia significativa en el precio promedio de productos entre empleados calificados y menos calificados.

```python
def prueba_hipotesis(*args):
    
    # Tenemos que mirar si las varianzas son iguales o no
    if len(args) == 2:
        p_valor_varianza = stats.levene(*args, center = "median")[1]
    else:
        p_valor_varianza = stats.bartlett(*args)[1]
    
    if p_valor_varianza > 0.05:
        # realizamos la prueba t de Student
        t_stat, p_valor = stats.ttest_ind(*args, equal_var=True)
    else:
        t_stat, p_valor = stats.ttest_ind(*args, equal_var=False)
        
    # Establecemos un nivel de significancia (alfa)
    alfa = 0.05

    # Comparamos el p-valor con el nivel de significancia
    if p_valor < alfa:
        print("Rechazamos la hipótesis nula.")
        print("Hay una diferencia significativa en los ingresos promedio entre los dos grupos.")
    else:
        print("No podemos rechazar la hipótesis nula.")
        print("No hay evidencia suficiente para afirmar una diferencia significativa en los ingresos promedio entre los dos grupos.")
```

```python
# Prueba de Hipótesis sobre el precio promedio de productos con o sin empleados calificados
df["employee_rat_cat"] = np.where(df["rating_employee"] > 4, "calificado", "no-calificado")
df_calificado = df[df["employee_rat_cat"] == "calificado"]["mean_product_prices"]
df_no_calificado = df[df["employee_rat_cat"] == "no-calificado"]["mean_product_prices"]
```

```python
prueba_hipotesis(df_calificado, df_no_calificado)
```
