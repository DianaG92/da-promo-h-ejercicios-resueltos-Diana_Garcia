EJERCICIOS DE GESTIÓN DE NULOS

1. Cargad el csv que obtuvimos en los ejercicios de la clase invertida de ayer. 

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
df_unido = pd.read_csv("rating_insights1.csv", index_col=0)

df_unido.head(2)
```

2. ¿Qué columnas tienen nulos? 

```python
columnas_nulas = df_unido[df_unido.columns[df_unido.isnull().any()]].select_dtypes(include = np.number).columns

print("Las columnas numéricas que tienen nulos son : \n ")
print(columnas_nulas)
```

3. Explorad los nulos que hay en el DataFrame. 

```python
df_unido[columnas_nulas].isnull().sum()
```

```python
df_unido[columnas_nulas].isnull().sum() / df_unido.shape[0]
```

4. Con lo aprendido en la lección de hoy gestiona los nulos de la forma que consideres más correcta. Mañana en clase discutiremos sobre los métodos elegidos. 

```python
# Sacamos el "value_counts()"" de cada una de las columnas categóricas que tienen nulos para saber cómo es la distribución de sus categorías.
for col in columnas_nulas:
    print(f"La distribución de las categorías para la columna {col.upper()}")
    display(df_unido[col].value_counts() / df_unido.shape[0])
    print("....................................................")
```

5. Guardad el csv para los ejercicios de la clase invertida de mañana con las columnas sin nulos. 

```python
df_unido = df_unido.dropna()
df_unido = df_unido.dropna(axis=1)

df_unido.head()
```
